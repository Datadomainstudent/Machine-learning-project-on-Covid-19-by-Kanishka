```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
import duckdb
conn=duckdb.connect()
df = pd.read_csv('corona_tested_006.csv')
```

```python
conn.register("df",df)
```

```python
conn.execute("select * from df").fetchdf()
```

```python
#1 Find the number of COVID patients who faced shortness of breath:

conn.execute("SELECT COUNT(*) AS NumPatientsWithShortnessOfBreath FROM df WHERE Corona = 'positive' AND Shortness_of_breath = true;").fetchdf()
```

```python
# 2. Find the number of negative COVID patients with fever and sore throat:

conn.execute("SELECT COUNT(*) AS NumNegativePatientsWithFeverAndSoreThroat FROM df where Corona = 'negative' AND Fever = true AND Sore_throat = true; ").fetchdf()
```

```python
# 3.Group the data by month and rank the number of positive cases:

conn.execute("SELECT MONTH(Test_date) AS Month, COUNT(*) AS NumPositiveCases FROM df WHERE Corona = 'positive' GROUP BY Month ORDER BY NumPositiveCases DESC;").fetchdf()
```

```python
# 4. Identify female negative COVID patients with cough and headache:

conn.execute("SELECT COUNT(*) AS NumFemaleNegativePatientsWithCoughAndHeadache FROM df WHERE Corona = 'negative' AND Sex = 'female' AND Cough_symptoms = true AND Headache = true;").fetchdf()
```

```python
# 5 Determine how many elderly COVID patients faced breathing problems:

conn.execute("SELECT COUNT(*) AS NumElderlyPatientsWithBreathingProblems FROM df WHERE corona = 'positive' AND Age_60_above = true AND Shortness_of_breath = true;").fetchdf()
```

```python
# 6 Identify the three most common symptoms among COVID-positive patients:

conn.execute("SELECT COUNT(*) AS SymptomCount, 'Cough_symptoms' AS Symptom FROM df WHERE Corona = 'positive' AND Cough_symptoms = trueUNION SELECT COUNT(*) AS SymptomCount, 'Fever' AS Symptom FROM df WHERE Corona = 'positive' AND Fever = true UNION SELECT COUNT(*) AS SymptomCount, 'Sore_throat' AS Symptom FROM df WHERE Corona = 'positive' AND Sore_throat = true ORDER BY SymptomCount DESC LIMIT 3; ").fetchdf()
```

```python
# 7 Identify the least common symptom among COVID-negative individuals:

conn.execute("SELECT COUNT(*) AS SymptomCount, 'Shortness_of_breath' AS Symptom FROM df WHERE Corona = 'negative' AND Shortness_of_breath = true UNION SELECT COUNT(*) AS SymptomCount, 'Headache' AS Symptom FROM df WHERE Corona = 'negative' AND Headache = true ORDER BY SymptomCount LIMIT 1; ").fetchdf()
```
